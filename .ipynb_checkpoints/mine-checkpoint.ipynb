```python
# General ML Packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# For dealing with imbalanced dataset
from imblearn.over_sampling import SMOTE

# Preprocessing 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder

# ML algorithm
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier

# Evaluation metrics
from sklearn.metrics import accuracy_score, classification_report, roc_curve, auc, RocCurveDisplay

# Saving the model
import joblib
```

```python
def generate_pie_plot(col: str):
    df[col].value_counts().plot.pie(autopct="%1.1f%%", colors=["skyblue", "salmon"])
    plt.title(f'{col} Proportion')
    plt.ylabel(None)
    plt.show()
```

```python
generate_pie_plot(col='PCOS_Diagnosis')
```

```python
# resampling our data 
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

smt = SMOTE(sampling_strategy='minority')
X_resample, y_resample = smt.fit_resample(X, y,)
y_resample
```

```python
y_resample.value_counts().plot.pie(autopct="%1.1f%%", colors=["skyblue", "salmon"])
plt.show()
```

```python
y_resample = LabelEncoder().fit_transform(y_resample)
X_resample = X_resample.values
X_resample.shape
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit(X_resample)


X_scaled = scaler.transform(X_resample)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_resample, test_size=.2, random_state=26)

print(X_train.shape,
    X_test.shape,
    y_train.shape,
    y_test.shape,
     )
```

```python
algorithms = {
    "LR": LogisticRegression(),
    "XGB": XGBClassifier(n_estimators=100,),
    "RF": RandomForestClassifier(n_estimators=100),
    "GB": GradientBoostingClassifier(),
    "AdaB": AdaBoostClassifier(),
    "SVM": SVC(),
}
accuracy_results = {}
for name, model in algorithms.items():
    m = model.fit(X_train, y_train)
    y_pred = m.predict(X_test)
    accuracy_results[name] = accuracy_score(y_pred, y_test)
```

```python
for name, score in accuracy_results.items():
    print(f'{name}: {score:.3}')
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=list(accuracy_results.keys()), y=list(accuracy_results.values()), palette="coolwarm")
plt.xlabel("ML Algorithms")
plt.ylabel("Accuracy Score")
plt.title("Model Accuracy Comparison for PCOS Diagnosis")
plt.xticks(rotation=30)
plt.ylim(0, 1)
plt.show()
```

```python
# So the result shown that our best model is XGBoost lets see the result of learning
model = XGBClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(classification_report(y_pred, y_test))
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)
display = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc,
                                  estimator_name='XGBoost')
display.plot()
plt.show()
```

```python
# Save the model
joblib.dump(model, 'model.pkl')
```
